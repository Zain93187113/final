# **Handling Null Values**

```python

selected_countries = ['China', 'India', 'United States']
years_of_interest = [str(year) for year in range(2000, 2021)]  # Years from 2000 to 2020


# PM2.5 Pollution Dataframe: Replace NaN with row mean for years 2000-2020
for year in years_of_interest:
    pm25_data[year] = pm25_data[year].fillna(pm25_data[years_of_interest].mean(axis=1))

# Urban Population Dataframe: Replace NaN with row mean for years 2000-2020
for year in years_of_interest:
    urban_pop_data[year] = urban_pop_data[year].fillna(urban_pop_data[years_of_interest].mean(axis=1))

# Verify the replacement by checking if there are any NaN values left for the years of interest
pm25_nan_check = pm25_data[years_of_interest].isnull().sum().sum() == 0
urban_pop_nan_check = urban_pop_data[years_of_interest].isnull().sum().sum() == 0

(pm25_nan_check, urban_pop_nan_check)
# Updating the approach to handle rows with all NaN values.

# Function to calculate the mean for a row, ignoring NaN values
def row_mean(row):
    # Use numpy to calculate mean, ignoring NaNs
    return row.dropna().mean()

# PM2.5 Pollution Data: Replace NaN with row mean for years 2000-2020, if available
for index, row in pm25_data[years_of_interest].iterrows():
    mean_val = row_mean(row)
    if pd.notna(mean_val):
        pm25_data.loc[index, years_of_interest] = row.fillna(mean_val)

# Urban Population Data: Replace NaN with row mean for years 2000-2020, if available
for index, row in urban_pop_data[years_of_interest].iterrows():
    mean_val = row_mean(row)
    if pd.notna(mean_val):
        urban_pop_data.loc[index, years_of_interest] = row.fillna(mean_val)

# Verify the replacement by checking if there are any NaN values left for the years of interest
pm25_nan_check = pm25_data[years_of_interest].isnull().sum().sum() == 0
urban_pop_nan_check = urban_pop_data[years_of_interest].isnull().sum().sum() == 0

(pm25_nan_check, urban_pop_nan_check)

```

# **Urban Population Trends (2000-2020)**

```python

import seaborn as sns
import matplotlib.pyplot as plt



urban_pop_filtered = urban_pop_data[['Country Name'] + years_of_interest]

urban_pop_filtered = urban_pop_filtered[urban_pop_filtered['Country Name'].isin(selected_countries)]
urban_pop_long = urban_pop_filtered.melt(id_vars='Country Name', var_name='Year', value_name='Urban Population (%)')

# Setting the plot size and style
sns.set(style="whitegrid")
# plt.figure(figsize=(10, 6))

# Plotting PM2.5 Pollution


plt.figure(figsize=(12, 6))
sns.lineplot(data=urban_pop_long, x='Year', y='Urban Population (%)', hue='Country Name')
plt.title('Urban Population Trends (2000-2020)')
plt.xticks(rotation=45)
plt.show()
```

# **Average PM2.5 Pollution Levels**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Assuming data is loaded into pandas dataframes: pm25_data and urban_pop_data

# Data Preparation
selected_countries = ['China', 'India', 'United States']
years_of_interest = [str(year) for year in range(2000, 2021)]  # Years from 2000 to 2020

pm25_filtered = pm25_data[['Country Name'] + years_of_interest]
pm25_filtered = pm25_filtered[pm25_filtered['Country Name'].isin(selected_countries)]
pm25_long = pm25_filtered.melt(id_vars='Country Name', var_name='Year', value_name='PM2.5 Pollution (µg/m³)')


# plt.figure(figsize=(12, 6))
sns.barplot(data=pm25_long, x='Country Name', y='PM2.5 Pollution (µg/m³)', estimator=np.mean, errorbar=None)
plt.title('Average PM2.5 Pollution Levels')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# **Pie Chart for Share of Urban and PM2.5 population**

```python

# Data Preparation
selected_countries = ['China', 'India', 'United States']
years_of_interest = [str(year) for year in range(2000, 2021)]  # Years from 2000 to 2020

pm25_filtered = pm25_data[['Country Name'] + years_of_interest]
pm25_filtered = pm25_filtered[pm25_filtered['Country Name'].isin(selected_countries)]
pm25_avg = pm25_filtered.drop(columns=['Country Name']).mean(axis=1)

urban_pop_filtered = urban_pop_data[['Country Name'] + years_of_interest]
urban_pop_filtered = urban_pop_filtered[urban_pop_filtered['Country Name'].isin(selected_countries)]
urban_pop_latest = urban_pop_filtered.iloc[:, -1]

# Pie chart for Urban Population
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.pie(urban_pop_latest, labels=urban_pop_filtered['Country Name'], autopct='%1.1f%%')
plt.title('Share of Urban Population in 2020')

# Pie chart for Average PM2.5 Pollution
plt.subplot(1, 2, 2)
plt.pie(pm25_avg, labels=pm25_filtered['Country Name'], autopct='%1.1f%%')
plt.title('Share of Average PM2.5 Pollution (2000-2020)')

plt.tight_layout()
plt.show()
```
